```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns; sns.set;

```

```python
df=pd.read_csv("Salary_dataset.csv")
df
```

```python
years=df["YearsExperience"]
salary=df["Salary"]

#values are already sorted in the data, which is a good thing!

plt.plot(years,salary)
plt.show()
```

```python

years = np.array(years)
years = years.reshape(-1, 1) #the model requires 2d data, so the reshape is necessary



model = LinearRegression() #object to the linear regression class
model.fit(years, salary) #this line does all the magic, here it implements the regression function and returns the result into the model, inplace


slope = model.coef_[0] #the result coefficient 
intercept = model.intercept_ #this is the y intercept


predicted_salary = model.predict(years)


plt.scatter(years, salary)
plt.plot(years, predicted_salary, color='yellow')


plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')


plt.grid(True)
plt.show()
```

```python
print(f"and this is the equation of the line: y={round(slope,2)}x + {round(intercept,2)}")
```
